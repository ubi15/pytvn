```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Calcolo numero di eventi

```python
eventi_t = []
utenti_t = []

cartella = '../../data/APS/aff_data_ISI_original_divided_per_month/'

utenti = set()
for fn in sorted(os.listdir(cartella)):
    lnum = 0
    with open(os.path.join(cartella, fn), 'rb') as f:
        for l in f:
            lnum += 1
            v = l.strip().split()
            for val in v:
                utenti.add(val)
            
    eventi_t.append(lnum)
    utenti_t.append(len(utenti))
```

```python
plt.plot(utenti_t, 'ob')
plt.loglog(eventi_t, 'sr')
```

```python
plt.plot(utenti_t, eventi_t)
#plt.plot(utenti_t, np.array(utenti_t))
```
